```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
#a)
x_support = [0, np.pi/3, np.pi/2]
y_support = [np.sin(0), np.sin(np.pi/3), np.sin(np.pi/2)]


def lagrange_interpolator(x, x_support, y_support):
    n = len(x_support)
    result = 0
    for i in range(n):
        term = y_support[i]
        for j in range(n):
            if i != j:
                term *= (x - x_support[j]) / (x_support[i] - x_support[j])
        result += term
    return result


x_vals = np.linspace(0, np.pi/2, 100)
y_vals = [lagrange_interpolator(x, x_support, y_support) for x in x_vals]

#b
plt.plot(x_vals, y_vals, label="Polinomio Interpolador", color="blue")
plt.scatter(x_support, y_support, color="red", label="Conjunto de Soporte", zorder=5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación de Lagrange con conjunto de soporte")
plt.legend()
plt.grid(True)
plt.show()

#c
x_estimate = np.pi / 8
sin_estimated = lagrange_interpolator(x_estimate, x_support, y_support)


sin_true = np.sin(x_estimate)

#d
error = abs(sin_true - sin_estimated)
#e
f_third_derivative = np.cos(x_estimate)
product_term = np.prod([x_estimate - x_i for x_i in x_support])
factorial_term = math.factorial(3)
error_theoretical = abs(f_third_derivative * product_term / factorial_term)

print(f"Valor estimado de sin(pi/8): {sin_estimated}")
print(f"Valor real de sin(pi/8): {sin_true}")
print(f"Error en la estimación: {error}")
print(f"Error teórico: {error_theoretical}")
print(f"Diferencia de errores: {error-error_theoretical}")
```

Como se ve ambos errores son de la misma magnitud asi que podemos asumir que son consistentes
