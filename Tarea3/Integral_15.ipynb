```python
import sympy as sp


x = sp.symbols('x')


zeros_list = []

def zeros_legendre(n):

    for n in range(1, n):
       Pn = sp.legendre(n, x)
       zeros = sp.solveset(Pn, x, domain=sp.Interval(-1, 1))
       zeros_list.append(list(zeros))
    return zeros_list

weights_list = []

def weights_legendre(n):
    for n in range(1, n):
        Pn = sp.legendre(n, x)
        Pn_prime = Pn.diff(x)


        zeros = sp.solveset(Pn, x, domain=sp.Interval(-1, 1))


        weights = []
        for zero in zeros:
            weight = 2 / ((1 - zero**2) * (Pn_prime.subs(x, zero))**2)
            weights.append(weight)

    return weights_list.append(weights)
z=zeros_legendre(10)
print(z)
w=weights_legendre(10)
print(w)
```

```python
import numpy as np
from scipy.special import roots_legendre


n_polynomials = 20


zeros_list = []
weights_list = []

for n in range(1, n_polynomials + 1):

    zeros, weights = roots_legendre(n)


    zeros_list.append(zeros.tolist())
    weights_list.append(weights.tolist())


for i in range(n_polynomials):
    print(f"Polinomio de Legendre de grado {i + 1}:")
    print(f"Ceros: {zeros_list[i]}")
    print(f"Pesos: {weights_list[i]}")
    print()
```
