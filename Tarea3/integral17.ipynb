```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp_special
# Puntos y pesos de la cuadratura de Gauss-Laguerre para n=3
Raices, Pesos=np.polynomial.laguerre.laggauss(3)
Raices, Pesos

# Definir la función g(x) = x^3 * e^x / (e^x - 1)
def g(x):
    return x**3 * (np.exp(x) / (np.exp(x) - 1))

# Calcular la aproximación usando la fórmula de cuadratura de Gauss-Laguerre
I_approx = sum(w * g(x) for x, w in zip(Raices, Pesos))
real=np.pi**4/15
print(real)

# Resultado aproximado
print(I_approx)

n_values = np.arange(2, 11)  # Valores de n desde 2 hasta 10
errors_relative = []

for n in n_values:
    # Obtener los puntos y pesos de Gauss-Laguerre para n puntos
    Raices, Pesos = sp_special.l_roots(n)

    # Aproximar la integral
    I_approx = sum(w * g(x) for x, w in zip(Raices, Pesos))

    # Calcular el error relativo
    epsilon_r = I_approx / real
    errors_relative.append(epsilon_r)

# Graficar el error relativo en función de n
plt.plot(n_values, errors_relative, marker='o', linestyle='-', color='b', label=r'$\epsilon_r(n)$')
plt.axhline(y=1, color='r', linestyle='--', label='Exacto')
plt.xlabel('Número de puntos de evaluación (n)')
plt.ylabel(r'$\epsilon_r(n)$')
plt.title('Error relativo en función de los puntos de evaluación')
plt.legend()
plt.grid(True)
plt.show()
```
