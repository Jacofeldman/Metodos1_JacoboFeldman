```python
import numpy as np
import math
import matplotlib.pyplot as plt

# MÃ©todo del trapecio
def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    y = f(x)
    integral = h * (0.5 * y[0] + 0.5 * y[-1] + np.sum(y[1:-1]))
    return integral,x,y


f = lambda x: np.exp(-x**2)

error_tolerance = 0.005
f_prime_max_value = 2
a = 0
b = 1


n_min = math.sqrt(((b - a)**3 * f_prime_max_value) / (12 * error_tolerance))


n_required = math.ceil(n_min)


integral_value,x,y = trapezoidal_rule(f, a, b, n_required)

print(f"NÃºmero de puntos necesarios (n): {n_required}")
print(f"Valor aproximado de la integral: {integral_value:.5f}")
```
