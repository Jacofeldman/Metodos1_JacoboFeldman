```python
data = pd.read_csv(file,sep=' ')
data.columns = ['x','y']
data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import lstsq
# Matriz de diseño para el ajuste a la forma f(x) = a0 + a1 * x
x = np.array(data['x'])
y = np.array(data['y'])
A = np.vstack([np.ones(len(x)), x]).T
coeffs, _, _, _ = lstsq(A, y)  # a0 y a1 son los parámetros encontrados
a0, a1 = coeffs

# Calcular la recta de ajuste
y_fit = a0 + a1 * x

# Graficar datos y ajuste
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Datos')
plt.plot(x, y_fit, color='red', label=f'Ajuste: $y = {a0:.2f} + {a1:.2f}x$')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Ajuste por Mínimos Cuadrados a la Forma $f(x) = a0 + a1x$")
plt.legend()
plt.grid(True)
plt.show()

a0, a1
```
