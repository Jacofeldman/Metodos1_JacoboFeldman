```python
import numpy as np
import matplotlib.pyplot as plt
# Función para calcular el ajuste por mínimos cuadrados de forma explícita
def linear_least_squares(x, y):
    l = x.shape[0]
    b = y
    # Matriz de diseño A para un ajuste lineal f(x) = a0 + a1 * x
    A = np.ones((l, 2))
    A[:, 1] = x  # Segunda columna contiene los valores de x

    # Calculamos la solución por mínimos cuadrados
    AT = np.dot(A.T, A)  # Producto de A transpuesta por A
    bT = np.dot(A.T, b)  # Producto de A transpuesta por b
    coeffs = np.linalg.solve(AT, bT)  # Resolvemos el sistema lineal
    return coeffs

# Aplicamos la función para obtener los coeficientes de ajuste lineal
a0, a1 = linear_least_squares(x_data_linear, y_data_linear)

# Calcular la recta de ajuste
y_fit = a0 + a1 * x_data_linear

# Graficar los datos y el ajuste lineal
plt.figure(figsize=(8, 6))
plt.scatter(x_data_linear, y_data_linear, color='blue', label='Datos')
plt.plot(x_data_linear, y_fit, color='red', label=f'Ajuste: $y = {a0:.2f} + {a1:.2f}x$')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Ajuste por Mínimos Cuadrados a la Forma $f(x) = a0 + a1x$")
plt.legend()
plt.grid(True)
plt.show()

a0, a1
```
