```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la espiral de Arquímedes
a = 0
b = 1

# Valores de theta en el rango [0, 2π]
theta = np.linspace(0, 2*np.pi, 1000)

# Cálculo de r para la espiral de Arquímedes
r = a + b * theta

# Conversión a coordenadas cartesianas
x = r * np.cos(theta)
y = r * np.sin(theta)

# Graficar la espiral
plt.figure(figsize=(6,6))
plt.plot(x, y)
plt.title("Espiral de Arquímedes (a=0, b=1)")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.grid(True)
plt.show()
```
