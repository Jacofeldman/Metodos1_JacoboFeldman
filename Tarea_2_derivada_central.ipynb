```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función f(x)
def f(x):
    return 1 / np.sqrt(1 + np.exp(-x**2))

# Derivada central
def derivada_central(f, x, h=0.05):
    return (f(x + h) - f(x - h)) / (2 * h)

# Derivada exacta (para cálculo de error)
def derivada_exacta(x):
    exp_term = np.exp(-x**2)
    return (x * exp_term) / ((1 + exp_term)**(3/2))

# Intervalo de x y el paso h
x = np.linspace(-10, 10, 1000)
h = 0.05

# Calcular la derivada estimada usando la derivada central
derivada_estimada = derivada_central(f, x, h)

# Calcular la derivada exacta
derivada_real = derivada_exacta(x)

# Calcular el error en cada punto nodal
error = np.abs(derivada_estimada - derivada_real)
conteo=0
for i in error:
  conteo=+i

a=conteo/len(error)
print(a)

# Graficar la derivada estimada y la derivada exacta
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, derivada_estimada, label="Derivada Estimada (Central)")
plt.xlabel("x")
plt.ylabel("Derivada")
plt.title("Derivada Estimada")
plt.legend()
plt.grid(True)
plt.subplot(1, 2, 2)
plt.plot(x, derivada_real, label="Derivada Exacta", linestyle='--')
plt.xlabel("x")
plt.ylabel("Derivada")
plt.title("Derivada Exacta")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
