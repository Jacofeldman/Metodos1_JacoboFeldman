```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return x**3

# Intervalo
a = 0
b = 2

# (e) Aproximación de la integral para n = 30
n = 30

def suma_riemann(f, a, b, n):
    x = np.linspace(a, b, int(n), endpoint=False)
    dx = (b - a) / n
    return np.sum(f(x)) * dx
riemann_sum=suma_riemann(f, a, b, n)
print(f'Aproximación de la integral para n={n}: {riemann_sum}')

# (f) Calcular y dibujar el error de la estimación
exact_value = (b**4) / 4  # Valor exacto de la integral
n_values = np.linspace(30, 400, 100, endpoint=False)
errors = []

for n in n_values:
    riemann_sum1 = suma_riemann(f, a, b, int(n))
    error = abs(exact_value - riemann_sum1)
    errors.append(error)

# Graficar el error
plt.figure(figsize=(10, 6))
plt.plot(n_values, errors, label='Error de la estimación', color='blue')
plt.xlabel('Número de subintervalos (n)')
plt.ylabel('Error')
plt.title('Error de la Estimación de la Integral usando Suma de Riemann')
plt.grid()
plt.legend()
plt.show()
```

g)la suma de Riemann es una buena estrategia para estimar integrales simples, cuando se busca alta precisión o cuando se trabaja con funciones complejas no lo es. Para aplicaciones que requieren una mayor exactitud o que involucran funciones complicadas, es preferible optar por métodos avanzados como la cuadratura de Gauss.
